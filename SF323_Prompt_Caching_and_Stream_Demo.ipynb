## Prompt Caching

```python
from google.colab import userdata
from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=userdata.get('openrouter'),
)

response = client.chat.completions.create(
      model='google/gemini-2.5-flash',
      messages=[{'role':'user','content':'Hello'}],
      temperature = 0,
)
print(response.choices[0].message.content)
```

```python
story = """Title: The Windcaller's Gambit

The morning mist clung to the Celestial Downs like a silk shroud, obscuring the massive crystal spires that marked each furlong of the track. Kira Stormwind pressed her hand against her mount's neck, feeling the electric thrum of magic beneath the midnight-blue coat. Zephyr wasn't just any racing horse—he was a Skydancer, one of the rare breeds whose hooves could find purchase on air itself.

"Easy, boy," she whispered, though her own heart hammered against her ribs. Today was the Equinox Stakes, the most prestigious race in all of Aethermoor. Winners became legends. Losers... well, some never raced again.

Around them, the other competitors made their final preparations. To her left, Lord Blackthorne sat atop his Shadowmere, a horse that could phase through solid matter for brief moments. The beast's coat seemed to absorb light, creating an unsettling void in the shape of a horse. To her right, the favorite—Duchess Irena Goldleaf on her Solarflare, a magnificent creature whose mane literally burned with harmless golden flames that left trails of light in their wake.

Kira was the outsider here, the only commoner in a field of nobility. She'd earned her spot through the qualifying races, where Zephyr's ability to run on wind currents had shocked the racing establishment. But the Equinox Stakes was different. Here, the track itself was alive with ancient magic.

The crystal spires began to hum, their resonance growing louder as the sun climbed toward its zenith. Master of Ceremonies Aldrick Timekeeper raised his staff, the chronometer at its tip spinning wildly.

"Riders!" his voice boomed across the Downs, magically amplified. "The Equinox approaches! You know the rules—three laps around the Celestial Circuit. The track will shift with each lap. Magic is permitted, but interference with another rider's mount is grounds for disqualification. May the swiftest prevail!"

Kira tightened her grip on Zephyr's reins—not leather, but woven from cloudsilk, the only material that could properly channel a Windcaller's power. She was one of the few remaining Windcallers, humans who could speak to the air itself and convince it to carry them. It was a dying art, one her grandmother had taught her in secret.

The other riders were already positioning themselves. She noticed Prince Caspian on his Tidereaver, a horse that could run across water as if it were solid ground. The track's second lap would take them across Lake Mirrors—a significant advantage for him. Behind them, the mysterious Lady Nightwhisper sat astride her Dreamwalker, a ethereal mount that seemed to exist partially in another realm.

"Steady," Kira murmured, as much to herself as to Zephyr. The horse snorted, pawing at the ground with one hoof while the other three stood on small platforms of condensed air.

The chronometer's spinning slowed. In moments, when the sun hit its exact zenith, the race would begin. Kira closed her eyes, reaching out with her senses to feel the wind patterns around the track. There—a strong current rising from the Thermal Gardens on the eastern curve. If she could catch that...

CRACK!

The starting crystal shattered in a burst of rainbow light, and twenty of the finest magical horses in the realm surged forward. The crowd's roar was deafening, but Kira heard only the wind singing in her ears.

Zephyr launched himself upward immediately, his hooves finding purchase on solid air as they rose above the pack. Several horses had the same idea—Lady Goldleaf's Solarflare ran on platforms of hardened light, while a young viscount's Stormbringer galloped across crackling bridges of lightning.

But Kira had studied the wind patterns for weeks. She guided Zephyr toward a powerful air current that would carry them around the first turn with minimal effort. Below, the earthbound horses—those whose magic was more subtle—took the traditional route. Lord Blackthorne's Shadowmere phased through other horses with disturbing ease, while a baroness's Ironhoof left deep gouges in the track, its metallic legs pumping like pistons.

The first lap was about establishing position. Kira and Zephyr swooped low as they approached the Serpent's Spine, a series of sharp turns that had ended many races. Here, the earthbound horses had an advantage—solid ground meant better control. But Kira had practiced this.

"Now!" she cried, and Zephyr folded his legs, diving like a falcon. They shot between two massive crystals just as Prince Caspian's Tidereaver created a water spout to propel himself up and over the turns entirely. Clever, but costly—she could see the strain on both horse and rider.

As they completed the first lap, the track shimmered and changed. The solid ground of the home stretch transformed into a series of floating platforms, while the eastern section became a maze of vertical walls. The crowd gasped as Lake Mirrors rose into the air, forming a spherical mass of water that the horses would have to navigate through.

This was where breeding and training truly mattered. A duke's Gravitydefier simply inverted its personal gravity and ran along the underside of the platforms. The Shadowmere flickered in and out of existence, taking shortcuts through the shadow realm. But Zephyr—Zephyr was in his element now.

Kira sang to the wind in the old tongue, words her grandmother had made her memorize until they were carved into her soul. The air responded, forming a tunnel of calm through the chaos. They shot through the floating lake, Zephyr's wind-shield parting the water like curtains. Behind them, she heard a crash—someone hadn't made it through cleanly.

By the time they started the third lap, the field had thinned. Only eight horses remained in true contention. Lady Goldleaf's Solarflare had taken the lead, leaving a trail of fire that made it dangerous for others to follow too closely. Lord Blackthorne was second, his Shadowmere's phasing ability allowing him to take impossible racing lines.

But Kira had saved their strength for this final lap. As the track transformed for the last time—becoming a twisted ribbon that spiraled up into the sky itself—she leaned forward and whispered the most powerful windword she knew.

Zephyr's neigh was like thunder as hurricane winds enveloped them. They didn't just run now—they flew, truly flew, abandoning any pretense of touching the track. It was dangerous, technically legal but pushing the very limits of the rules. The other riders cried out in alarm as Kira and Zephyr shot past them, riding a wind that seemed to come from the heart of the storm itself.

Lady Goldleaf pushed her Solarflare harder, the horse's flames turning from gold to white-hot silver. Lord Blackthorne attempted to phase directly to the finish line, but the anti-cheating wards crackled, forcing his mount back into full materiality. The Dreamwalker tried to shift the race into a dream where it had already won, reality rippling around it like water.

But Zephyr and Kira had become one with the wind. They crossed the finish line a full length ahead, touching down just as the chronometer chimed the end of the race. The crowd erupted, a mix of cheers and shocked gasps. A commoner had won the Equinox Stakes. A Windcaller—one of the old magic users that the nobility had tried so hard to forget—had proven that the ancient ways still had power.

As Kira dismounted on shaking legs, Zephyr nuzzling her shoulder proudly, she saw her grandmother in the crowd, tears streaming down her weathered face. Lord Blackthorne approached, his face unreadable.

"That was... extraordinary," he said finally. "I've never seen anyone command the wind like that. Where did you learn?"

Kira straightened, meeting his eyes steadily. "From those who came before. From those who remembered that magic isn't just about breeding or gold—it's about the bond between rider and mount, between person and element."

The Duke nodded slowly, then did something unexpected—he bowed. "Then perhaps we nobles have something to learn from you, Windcaller Stormwind."

As the sun began its descent from zenith, painting the sky in brilliant oranges and purples, Kira stood with Zephyr in the winner's circle. The crystal trophy felt heavy in her hands, but not as heavy as the responsibility she now carried. She had proven that the old ways could compete with the new, that a commoner could stand equal to nobility.

But more than that, she had shown that in the world of fantasy horse racing, the greatest magic wasn't in the blood of legendary steeds or the coffers of the wealthy—it was in the partnership between human and horse, in the willingness to trust each other completely, and in the courage to ride the wind wherever it might lead.

The Equinox Stakes would be remembered not just for its winner, but for the day the wind itself returned to the Celestial Downs, carrying with it echoes of an older, wilder magic that refused to be forgotten.
"""

messages = [
  {
    "role": "system",
    "content": [
      {
        "type": "text",
        "text": "You are a helpful assistant to answer questions about the following fantasy story: "
      },
      {
        "type": "text",
        "text": story,
        "cache_control": { "type": "ephemeral" }  #Prompt break point
      }
    ]
  },
  {
    'role':'user',
    'content':'Please summarize the story'
  }
]
```

```python
response = client.chat.completions.create(
      model='google/gemini-2.5-flash', # only flash seems to work with explicit caching on OpenRouter
      messages=messages
)
print(response.choices[0].message.content)
```

## Streaming

- The OpenRouter API allows streaming responses from any model.

- This is useful for building chat interfaces or other applications where the UI should update as the model generates the response.

- To enable streaming, you can set the `stream` parameter to `true` in your request.

https://openrouter.ai/docs/api-reference/streaming#provider-support

https://platform.openai.com/docs/guides/streaming-responses

#### Streaming output

```python
stream = client.chat.completions.create(
      model='google/gemini-2.5-flash',
      messages=messages,
      stream = True
)

for event in stream:
    print(event)
```

```python
stream = client.chat.completions.create(
      model='google/gemini-2.5-flash',
      messages=[{'role':'user','content':'Write me an fantasy adventure story.'}],
      stream = True
)

for event in stream:
    print(event.choices[0].delta.content, end='')
```
